```python
def filter(paper : Paper):
    title = paper.title
    # Define common keywords related to speech synthesis, voice conversion, etc.
    keywords = [
        "to-speech", "to speech", "voice", "synthes", "conversion", "TTS", "style", "emotion", "accent",
        "prosody", "native", 
        "any-to-any"
    ]
    
    # Check if any of the keywords appear in any of the titles

    if any(keyword.lower() in title.lower() for keyword in keywords):
        return True
    return False
```

```python
pid_to_i = {node.paperId : i for i, node in enumerate(nodes)}
```

```python
node_dict = {node.paperId:node for node in nodes}
for node in nodes:
    node.out = []
    node.in_ = []

for node in nodes:
    if node.paper._citations is not None:
        for to in node.paper._citations:
            if to in node_dict:
                node.out.append(to)
                node_dict[to].in_.append(node.paperId)
    
    if node.paper._references is not None:       
        for from_ in node.paper._references:
            if from_ in node_dict:
                node.in_.append(from_)
                node_dict[from_].out.append(node.paperId)

for node in nodes:
    node.out = set(node.out)
    node.in_ = set(node.in_)              
```

```python
paper_list = [node.paperId for node in nodes]
```

```python
metric_median = np.median([p.metric() for p in Library.get(paper_list) 
                           if p.citationCount is not None],axis=0)
```

```python
def paper_score(pId, fn=np.mean):
    return fn(Library.get(pId).metric() / metric_median)
```
